## 1ª Questão:
Seja uma população com distribuição normal com variância σ² = 36.
Toma-se uma amostra aleatória de tamanho n = 16 e a média estimada
para a amostra foi de x = 43. Ao nível de confiância α = 90%, calcule o intervalo de confiança para essa média.

```python
# Definindo os valores
mediaAmostral = 43
desvio = 36**(1/2)
n = 16
numeros = np.random.normal(mediaAmostral,desvio,n)

# Agora plotando a distribuição dos dados
sns.distplot(numeros, hist=False,rug=False)
plt.title('Distribuição Normal')
plt.axvline(mediaAmostral,color='k')
plt.show()
```

```python
# Ao observar a tabela, obtemos os valores de zc e -zc (ambos 0.05):
zc = 1.645

# Agora que temos todos os dados, podemos calcular a margem de erro
margem = zc*(desvio/n**(1/2))

# Agora encontrando os extremos
extEsquerdo = mediaAmostral - margem
extDireito = mediaAmostral + margem
print("O intervalo de confiança é:",extEsquerdo,"<",mediaAmostral,"<",extDireito)
```

## 3ª Questão:

## 6ª Questão:
Um certo tipo de rato apresenta, nos três primeiros meses de vida, um ganho médio de peso de 58g. Uma amostra de 10 ratos foi alimentada desde o nascimento até a idade de 3 meses com uma ração especial, e o ganho de peso médio dos ratos foi de 55, 58, 60, 62, 65, 67, 54, 64, 62 e 68. Com uma probabilidade de 95%, o intervalo de confiança para essa amostra contém a média de 58g?

```python
# Definindo os valores

lista = [55, 58, 60, 62, 65, 67, 54, 64, 62, 68]

# Calculando a média

mediaAmostral = 0
for i in lista:
  mediaAmostral+=i
mediaAmostral/=len(lista)

# Calculando a variância

variancia = 0
for valor in lista:
  variancia += (valor - mediaAmostral)**2
variancia = round(variancia/len(lista),2)

# Descobrindo o Desvio Padrão
desvio = variancia**(1/2)
n = len(lista)
numeros = np.random.normal(mediaAmostral,desvio,n)

# Agora plotando a distribuição dos dados
sns.distplot(numeros, hist=False,rug=False)
plt.title('Distribuição Normal')
plt.axvline(mediaAmostral,color='k')
plt.show()
```

```python
# Ao observar a tabela t student, obtemos o valor do grau de liberdade (1.833):
# Agora que temos todos os dados, podemos calcular a margem de erro
margem = 1.833*(desvio/n**(1/2))

# Agora encontrando os extremos
extEsquerdo = round(mediaAmostral - margem,5)
extDireito = round(mediaAmostral + margem,5)
print("O intervalo de confiança é:",extEsquerdo,"<",mediaAmostral,"<",extDireito)
```
