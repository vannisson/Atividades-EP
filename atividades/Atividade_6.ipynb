## Correlação

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

url = "https://raw.githubusercontent.com/vannisson/Atividades-EP/main/datasets/deaths_brazil.csv"
aux = pd.read_csv(url,usecols=[0,3])
year = aux['year'].tolist()
deaths = aux['deaths'].tolist()

plt.scatter(year,deaths)
plt.xlabel('Ano')
plt.ylabel('Mortes')
plt.title('Diagrama de Dispersão')
plt.show()
```

```python
np.corrcoef(year,deaths)[0,1]
```

## Regressão

```python
from scipy import stats
m, b , r_value, p_value,std_err = stats.linregress(year,deaths)
m,b
```

```python
linha_reg = [m*i + b for i in range(max(deaths))]
plt.scatter(year,deaths)
plt.plot(linha_reg,'r')
plt.xlabel('Ano')
plt.ylabel('Mortes')
plt.show()
```
